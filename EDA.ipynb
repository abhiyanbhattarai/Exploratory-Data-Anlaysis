```python
def hist_box(data,xlabel=None, title=None, font_scale=2,figsize=(15,7),bins=None):
  mean=np.mean(data)
  sns.set(font_scale = font_scale) 
  f2, (ax_box2, ax_hist2) = plt.subplots(2, sharex = True, gridspec_kw = {"height_ratios": (.25, .75)}, figsize = figsize) 
  sns.boxplot(x = data, ax = ax_box2, showmeans = True, color = "violet")
  sns.histplot(x = data, kde = False, ax = ax_hist2, bins = bins, palette="winter") if bins else sns.histplot(x = data, kde = False, ax = ax_hist2, color = "lightblue")
  ax_hist2.axvline(mean, color = 'g', linestyle = '--') 
  if xlabel: ax_hist2.set(xlabel = xlabel) 
  if title: ax_box2.set(title = title) 
  plt.show()
```

**Total Score**

```python
hist_box(data['Total score'])
```

Student prefer standard over free/reduced.

**Average Score Vs Lunch Type**

```python
br=data.groupby('lunch')['Total score'].mean()
ax=br.plot(kind='bar', ylabel='Average Score', xlabel='Lunch Type',title='Average Score vs Lunch Type',figsize=(10, 6))
plt.xticks(rotation=0)
print('Average Score for',br)
```

Students having standard lunch scored more than free/reduced lunch.

**Average Score vs Race/Ethnicity**

```python
br=data.groupby('race/ethnicity')['Total score'].mean()
br.plot(kind='bar', ylabel='Average Score', xlabel='Race/Ethnicity',title='Average Score vs Race/Ethnicity', figsize=(10, 6))
plt.xticks(rotation=0)
print('Average Score for',br)
```

Group E students are better perfomer than others.

**Average Score Vs. Gender**

```python
br=data.groupby('gender')['Total score'].mean()
br.plot(kind='bar', ylabel='Average Score', xlabel='Gender',title='Average Score Vs. Gender', figsize=(10, 6))
plt.xticks(rotation=0)
print('Average Score for',br)
```
